```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

# 1. Model & Tokenizer load chestunnam
print("Model load avtundi, koncham wait cheyi...")
tokenizer = AutoTokenizer.from_pretrained("microsoft/DialoGPT-medium")
model = AutoModelForCausalLM.from_pretrained("microsoft/DialoGPT-medium")

print("Chatbot: Ready! Message type cheyi (exit kottu aapali ante).")

# 2. Chatting Loop
while True:
    user_input = input("You: ")
    if user_input.lower() in ['exit', 'quit', 'bye']:
        print("Chatbot: Bye! Malli kaludham.")
        break

    # Input process
    new_user_input_ids = tokenizer.encode(user_input + tokenizer.eos_token, return_tensors='pt')

    # AI Response generation
    chat_history_ids = model.generate(new_user_input_ids, max_length=1000, pad_token_id=tokenizer.eos_token_id)

    # Output decoding
    response = tokenizer.decode(chat_history_ids[:, new_user_input_ids.shape[-1]:][0], skip_special_tokens=True)
    print(f"Chatbot: {response}")
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

# Model ni ready chestunnam
tokenizer = AutoTokenizer.from_pretrained("microsoft/DialoGPT-medium")
model = AutoModelForCausalLM.from_pretrained("microsoft/DialoGPT-medium")

print("Chatbot Ready ayipoindi! 'Hi' ani type cheyi.")

while True:
    user_msg = input("You: ")
    if user_msg.lower() in ['exit', 'quit']: break

    inputs = tokenizer.encode(user_msg + tokenizer.eos_token, return_tensors='pt')
    outputs = model.generate(inputs, max_length=1000, pad_token_id=tokenizer.eos_token_id)
    response = tokenizer.decode(outputs[:, inputs.shape[-1]:][0], skip_special_tokens=True)

    print(f"Chatbot: {response}")
```
